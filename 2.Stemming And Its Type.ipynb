```python
## Classification Problem
## Comment of product is positive review or negative review
## Reviews ----> [eating, eaten ] ---> eat,[going, gone, goes] ---> go
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programms", "history", "finally", "finalize"]
```

```python
for word in words:
    print(word)
```

### **PorterStemmer**

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + " ----> " + stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

```python
stemming.stem("city")
```

```python
stemming.stem("sitting")
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|label$', min = 0)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
## Snowball Stem

from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word + " ---> " + snowball_stemmer.stem(word))
```

```python
stemming.stem('fairly'), stemming.stem('sportingly')
```

```python
snowball_stemmer.stem('fairly'), snowball_stemmer.stem('sportingly')
```

```python
snowball_stemmer.stem("going")
```

```python
snowball_stemmer.stem("goes")
```
